### map, reduce, filter


* `map (func, iterables)` - applies func to the elements of *iterables* and returns **new** `iterable`:

````python
words = dir(list)
````

````python
letter_counts = list(map(lambda x: len(x), words))

print(letter_counts)
print()
print(words)
````

````python
l1 = [1,2,3,4]
l2 = [11, 12, 13, 14, 15]
l3 = [101, 102, 103]

triple_sum = list(map(lambda x, y, z: x + y + z, l1, l2, l3))
print(triple_sum)
````


* `reduce (func, iterables)` - performs a calculation with the elements of a sequence, the result of which is **one value**:

````python
from functools import reduce
````

````python
sum_of_counts = reduce(lambda x, y: x + y, letter_counts)
print(sum_of_counts)
````


* `filter (predicate, iterable)` - leaves only those elements for which *the predicate* is **true** (function returning bool):

````python
mixed = ['honey', 'chocolate', 'honey', 'chocolate', 'chocolate', 'honey', 'chocolate', 'chocolate', 'honey', 'chocolate']
only_honey = list(filter(lambda x: x == 'honey', mixed))
print(only_honey)
````

| **'r'** | Opening to **read** (default) |

| **'w'** | Opening to **write**, the contents of the file is deleted. If the file does not exist, a new one is created |

| **'a'** | Opening to **write**, information is appended **at the end of the file** |
| **'r+'** | Open file to **read & write**. If there is no file, **a new one is NOT created** |
| **'a+'** | Open the file to **read and apend to the end of the file**. If there is no file, **new one is created** |
| **'t'** | Opening a file **as text** (default) |

### classes

````python
class Human:
    def __init__(self, name='', age=None, deep_learning_specialist=False):
        self.name = name
        self.age = age
        self.deep_learning_specialist = deep_learning_specialist
    
    def set_age(self, age):
        self.age = age
    
    def get_age(self):
        return self.age
    
    def __str__(self):
        return 'Name: {}\nAge: {} \
                \nIs deep learning specialist: {}'.format(self.name, self.age, 
                                                          self.deep_learning_specialist)

class DLSchoolStudent(Human):
    def __init__(self, name='', age=None, deep_learning_specialist=False):
        super().__init__(name, age, deep_learning_specialist)
        self.total_grade = None
        self.deep_learning_specialist = True
    
    def __str__(self):
        return super().__str__()
````

````python
human = Human('Person', 17)
print(human)
````

````python
student = DLSchoolStudent('Good Person', 18)
print(student)
````

### collections


* `defaultdict ()` - the dictionary class, which has a default value - sometimes can be handy:

````python
from collections import defaultdict
````

````python
d = defaultdict(int)

print(d['key'])

d['key'] = 5
print(d['key'])  # 5
````

````python
d = defaultdict(lambda: 'empty')
print(d['key'])

d['key'] = 'full'
print(d['key'])
````

````python
d = defaultdict(list)
print(d)

d['list1'].append(100)
d['list1'].append(200)
print(d)
print(d['list1'])
````
